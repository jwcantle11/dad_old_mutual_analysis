```python
#selenium (getting to the page that we will scrape)

#going to sign in page
driver.get('https://customer.oldmutualwealth.co.uk/Authentication')
userNameEnter = driver.find_element_by_class_name('input-validation-error')
username = 'acantle'

#promp user to enter username
driver.execute_script("var a = prompt('Enter username (10 seconds to enter before error)', '');document.body.setAttribute('data-id', a)")
time.sleep(10)
username = driver.find_element_by_tag_name('body').get_attribute('data-id')

#entering username
userNameEnter.clear()
userNameEnter.send_keys(username)
userNameEnter.send_keys(Keys.RETURN)

#prompt user to enter password
driver.execute_script("var a = prompt('Enter password (10 seconds to enter before error)', '');document.body.setAttribute('data-id', a)")
time.sleep(10)
password = driver.find_element_by_tag_name('body').get_attribute('data-id')

#entering password
passwordEnter = driver.find_element_by_class_name('input-validation-error')
passwordEnter.clear()
passwordEnter.send_keys(password)
passwordEnterButton = driver.find_element_by_id('signin')
passwordEnterButton.click()


#Scraping through each month from start 
#Current date: date which we are currently looking at. Today Date: the date the code was run

currentDate = datetime.datetime(2001, 6, 5)
todayDate = datetime.datetime.today()


while currentDate < todayDate:
    #getting day month and year
    day = currentDate.strftime("%d")
    month = currentDate.strftime("%m")
    year = currentDate.strftime("%Y")
    
    #changing due to british dates being different to python standard (US)
    dateToEnter = day + '/' + month + '/' + year
    dateEnterBox = driver.find_element_by_id('valuationdatepicker')

    #enter date into date box
    dateEnterBox.clear()
    dateEnterBox.send_keys(dateToEnter)
    dateEnterBox.send_keys(Keys.RETURN)
    #wait, as it takes a second to allow next element to show up
    time.sleep(2)
    
    #click on submit button to submit date
    dateSubmit = driver.find_element_by_class_name('gaptop-5')
    dateSubmit = dateSubmit.find_element_by_class_name('submit')
    dateSubmit.click()
    
    
    #getting html of page the driver is currently on
    pageHtml = BeautifulSoup(driver.page_source, 'html.parser')
    #call the scrape data
    americanStyleDate = month + '/' + day + '/' + year
    scrapeData(pageHtml, dateToEnter) #replace dateToEnter with americanStyleDate if you want date to be american in the dataset
    
    #go to next month
    currentDate = currentDate + relativedelta(months=+1)
    
```
